## Binarize label maps

nnUNet wants consecutive labels: 0, 1, etc.

Originally, the label map in the CHAOS challenge contains only two labels:
- 0 (background),
- 255 (region of interest).

We can simply binarize the label map.

```python
# !sudo git clone https://github.com/kevinkwshin/Keeplearning.git

import Keeplearning
from Keeplearning.list_processing import *
from Keeplearning.load_data import *

import glob

#################################################################
path_originalData = "/workspace/nnUnet/data/Task01_Tendon_5ch/"
# path_originalData = "/workspace/nnUnet/data/Task02_Tendon_4ch/"
# path_originalData = "/workspace/nnUnet/data/Task03_Tendon_6ch/"
# path_originalData = "/workspace/nnUnet/data/Task04_Tendon_1ch/"
# path_originalData = "/workspace/nnUnet/data/Task06_Pelvic/"
#################################################################


# list_sort_nicely < -- 이거 제 깃에서 가져오는 건데 정렬만해주는거라 필요없으시면 다른거로 대체하시죠
train_image = list_sort_nicely(glob.glob(path_originalData+"imagesTr/*"))
train_label = list_sort_nicely(glob.glob(path_originalData+"labelsTr/*"))
test_image = list_sort_nicely(glob.glob(path_originalData+"imagesTs/*"))
test_label = list_sort_nicely(glob.glob(path_originalData+"labelsTs/*"))
```

```python
train_image = [
             "{}".format(patient_no.split('/')[-1]) for patient_no in train_image
]

train_label = [
             "{}".format(patient_no.split('/')[-1]) for patient_no in train_label
]

test_image = [
             "{}".format(patient_no.split('/')[-1]) for patient_no in test_image
]

len(train_image),len(train_label),len(test_image),len(test_label), train_image[0]
```

# Change Batch size in plans

```python
import pickle
path_pickle = 'nnUNet_preprocessed/Task006_Pelvic/nnUNetPlansv2.1_plans_3D.pkl'

with open(path_pickle, 'rb') as handle:
    temp = pickle.load(handle)
    
print('Original',temp['plans_per_stage'][0]['batch_size'])
#############################################################################
temp['plans_per_stage'][0]['batch_size'] = 2  # here is the batch size
#############################################################################
print('After',temp['plans_per_stage'][0]['batch_size'])

with open(path_pickle, 'wb') as handle:
    pickle.dump(temp, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
