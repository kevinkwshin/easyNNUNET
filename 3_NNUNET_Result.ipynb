```python
def one_hot(np_array,channel,ignore_background=True):
    assert channel > 1
    
    results = []
    for idx in range(channel+1):
        
        result = np_array.copy()
        
        if idx == 0:
            if ignore_background == True:
                pass
            else:
                result[result!=0] = 2
                result[result==0] = 1
                result[result==2] = 0
                results.append(result)
                
        else:
            result[result!=idx] = 0
            result[result!=0] = 1
            results.append(result)
    
    return np.array(results)       
```

```python
dices = []
recalls = []
specificitys = []

for idx_id in range(len(list_pred)):
    num_ch = 6
    print('\n',idx_id,list_gt[idx_id],list_pred[idx_id],'\n')
    gt = data_load_nii(list_gt[idx_id])
    pred = data_load_nii(list_pred[idx_id])
    gt = one_hot(gt,num_ch)
    pred = one_hot(pred,num_ch)

    for idx_ch in range(num_ch):
        dice = score_dice(gt[idx_ch],pred[idx_ch])
        recall = score_recall(gt[idx_ch],pred[idx_ch])
        specificity = score_specificity(gt[idx_ch],pred[idx_ch])
#         print(idx_ch+1, dice)
        dices.append(dice)
        recalls.append(recall)
        specificitys.append(specificity)
```
